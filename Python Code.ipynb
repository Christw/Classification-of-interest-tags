```python
import pandas as pd
from google.colab import files
```

```python
# Read a file
df =pd.read_excel('Name of a file.xlsx')
df
```

```python
# Find out if there is missing data in the original file
df.isnull().value_counts()
```

```python
# Delete the segment with missing data 
df = df.dropna()
```

```python
# Reference of a classification
# Since some data are confidential, most of it is modified and doesn't fit the reality
new_third_tags_to_first = {
"3C":"3C產品",
"3C品牌":"3C產品",
"二手":"交通",
"品牌":"保養美體",
"大眾":"交通",
}
```

```python
# Mapping the data
# Examine if any data are mismatched after being classified 
df["new_seg"] = df.segment.map(new_third_tags_to_first)
df.isnull().value_counts()
```

```python
# Find out the missing mapping data
df[df.new_seg.isnull()]['segment'].unique()
```

```python
df
```

```python
new_df=df[['new_seg','f0_']]
new_df
```

```python
# Aggregate the needed data
group_count = new_df.groupby(['new_seg']).sum().sort_values(by=['f0_'],ascending=False)
group_count.columns=['counts']
group_count
```

```python
# Export the aggregated report and download it
group_count.to_excel("New Name.xlsx") 
files.download("New Name.xlsx")
```
